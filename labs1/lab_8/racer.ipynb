```python
# Импорттау
import pygame, sys
from pygame.locals import *
import random, time

# Pygame-ды инициализациялау
pygame.init()

# Кадр жиілігін орнату
FPS = 60
FramePerSec = pygame.time.Clock()

# Түстерді анықтау
BLUE  = (0, 0, 255)
RED   = (255, 0, 0)
GREEN = (0, 255, 0)
BLACK = (0, 0, 0)
WHITE = (255, 255, 255)

# Айнымалыларды орнату
WIDTH = 400
HEIGHT = 600
SPEED = 5
SCORE = 0

# Қаріптерді орнату
font = pygame.font.SysFont("Verdana", 60)
font_small = pygame.font.SysFont("Verdana", 20)
game_over = font.render("Game Over", True, BLACK)

# Фондық суретті жүктеу
background = pygame.image.load(r"C:\Users\nurga\OneDrive\Desktop\Projects\PP2\labs1\lab_8\images\AnimatedStreet.png")

# Ақ түсті экран жасау
screen = pygame.display.set_mode((400,600))
screen.fill(WHITE)
pygame.display.set_caption("Game")

# Ойын объектілерін жасау
class Enemy(pygame.sprite.Sprite):
      def __init__(self):
        super().__init__() 
        self.image = pygame.image.load(r"C:\Users\nurga\OneDrive\Desktop\Projects\PP2\labs1\lab_8\images\Enemy.png")
        self.rect = self.image.get_rect()
        self.rect.center = (random.randint(40, WIDTH-40), 0)

      def move(self):
        self.rect.move_ip(0,SPEED)
        if (self.rect.bottom > 600):
            self.rect.top = 0
            self.rect.center = (random.randint(40, WIDTH - 40), 0)

class Coin(pygame.sprite.Sprite):
    def __init__(self):
        super().__init__() 
        original_image = pygame.image.load(r"C:\Users\nurga\OneDrive\Desktop\Projects\PP2\labs1\lab_8\images\Coin.png")
        self.image = pygame.transform.scale(original_image, (30, 30))
        self.rect = self.image.get_rect()
        self.rect.center = (random.randint(40, WIDTH-40), 0)


    def move(self):
        self.rect.move_ip(0,SPEED)
        if (self.rect.bottom > 600):
            self.rect.top = 0
            self.rect.center = (random.randint(40, WIDTH - 40), 0)

    def reset(self):
        self.rect.top = 0
        self.rect.center = (random.randint(40, WIDTH - 40), 0)

class Player(pygame.sprite.Sprite):
    def __init__(self):
        super().__init__() 
        self.image = pygame.image.load(r"C:\Users\nurga\OneDrive\Desktop\Projects\PP2\labs1\lab_8\images\Player.png")
        self.rect = self.image.get_rect()
        self.rect.center = (160, 520)
       
    def move(self):
        pressed_keys = pygame.key.get_pressed()
        
        if self.rect.left > 0:
              if pressed_keys[K_LEFT]:
                  self.rect.move_ip(-5, 0)
        if self.rect.right < WIDTH:        
              if pressed_keys[K_RIGHT]:
                  self.rect.move_ip(5, 0)
                  

# Спрайттарды орнату        
P1 = Player()
E1 = Enemy()
C1 = Coin()

# Спрайт топтарын құру
enemies = pygame.sprite.Group()
enemies.add(E1)
friends = pygame.sprite.Group()
friends.add(C1)
all_sprites = pygame.sprite.Group()
all_sprites.add(P1)
all_sprites.add(E1)
all_sprites.add(C1)

# Ойын жылдамдығын арттыру үшін оқиға қосу
INC_SPEED = pygame.USEREVENT + 1
pygame.time.set_timer(INC_SPEED, 1000)

# Ойын циклі
while True:
      
    # Барлық оқиғаларды тексеру  
    for event in pygame.event.get():
        if event.type == INC_SPEED:
              SPEED += 0.5
        if event.type == QUIT:
            pygame.quit()
            sys.exit()

    # Фонды көрсету
    screen.blit(background, (0,0))
    scores = font_small.render(str(SCORE), True, BLACK)
    screen.blit(scores, (10,10))

    # Барлық спрайттарды жылжыту және қайта салу
    for entity in all_sprites:
        entity.move()
        screen.blit(entity.image, entity.rect)

    # Егер ойыншы жауға соқтығысса
    if pygame.sprite.spritecollideany(P1, enemies):
          time.sleep(1)
                    
          screen.fill(RED)
          screen.blit(game_over, (30,250))
          
          pygame.display.update()
          for entity in all_sprites:
                entity.kill() 
          time.sleep(2)
          pygame.quit()
          sys.exit()       
    
    # Егер ойыншы монетаны алса, ұпай қосылады
    elif pygame.sprite.spritecollideany(P1, friends):
        SCORE += 1

        for coin in friends:
            coin.reset()
        
    # Экранды жаңарту
    pygame.display.update()
    FramePerSec.tick(FPS)
```
