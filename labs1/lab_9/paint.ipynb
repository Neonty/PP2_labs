```python
import random
import pygame
import math

screen = pygame.display.set_mode((800, 600)) #экран өлшемін орнату
screen.fill((0, 0, 0)) #экранды қара түске бояу
work_surf = pygame.Surface((800, 600)) #сурет салуға арналған бет жасау
mode = 'random' #түстер режимі
draw_on = False #сурет салу күйін бақылау
last_pos = (0, 0) #соңғы позицияны сақтау
color = (255, 128, 0) #бастапқы түс
radius = 1 #сызық қалыңдығы
figure = 'pen' #салынатын фигура түрі

colors = { #түстер жинағы
    'red': (255, 0, 0),
    'blue': (0, 0, 255),
    'green': (0, 255, 0)
}

points = list() #көпбұрыштар үшін координаттар тізімі

def Rect_pos(x1, y1, x2, y2):  
    return pygame.Rect(min(x1, x2), min(y1, y2), abs(x1 - x2), abs(y1 - y2)) #тік төртбұрыш координаттарын есептеу

def Square_pos(x1, y1, x2, y2): 
    return pygame.Rect(min(x1, x2), min(y1, y2), abs(x1 - x2), abs(x1 - x2)) #шаршы координаттарын есептеу

def Equilateral_Triangle_pos(x1, y1, x2, y2): 
    base_length = abs(x2 - x1)
    
    height = (math.sqrt(3)/2) * base_length
    
    points.append((x1, y1))
    points.append((x2, y1))
    points.append(((x1+x2)//2, y1 - height))
    
    return points #тең қабырғалы үшбұрыш координаттарын есептеу

def Right_Triangle_pos(x1, y1, x2, y2): 
    points.append((x1, y1))
    points.append((x2, y2))
    points.append((x1, y2))
    return points #тік бұрышты үшбұрыш координаттарын есептеу

def Rhombus_pos(x1, y1, x2, y2): 
    points.append((x1, y1))
    points.append((x2, y2))
    points.append((x2-x1+x2, y1))
    points.append((x2, y1+y1-y2))
    return points #ромб координаттарын есептеу

running = True #ойын циклінің аяқталуын бақылау
while running:
    pressed = pygame.key.get_pressed() #пернетақта батырмаларын оқу

    for event in pygame.event.get():
        if event.type == pygame.QUIT:
            running = False #программаны жабу

        if event.type == pygame.KEYDOWN: #перне басу оқиғасы
            if event.key == pygame.K_1: 
                mode = 'red'
            if event.key == pygame.K_2: 
                mode = 'blue'
            if event.key == pygame.K_3: 
                mode = 'green'
            if event.key == pygame.K_q:
                figure = 'rectangle'
            if event.key == pygame.K_w: 
                figure = 'pen'
            if event.key == pygame.K_e: 
                figure = 'erase'
            if event.key == pygame.K_r: 
                figure = 'circle'
            if event.key == pygame.K_t: 
                figure = 'square'
            if event.key == pygame.K_y: 
                figure = 'equilateral triangle'
            if event.key == pygame.K_u: 
                figure = 'right triangle'
            if event.key == pygame.K_i: 
                figure = 'rhombus'

        if event.type == pygame.MOUSEBUTTONDOWN: #тышқан батырмасын басу
            if mode == 'random': 
                color = (random.randrange(256), random.randrange(256), random.randrange(256)) #кездейсоқ түс таңдау
            else:
                color = colors[mode] #таңдалған түсті орнату
            draw_on = True #сурет салуды бастау
            last_pos = event.pos #бастапқы позицияны сақтау

        if event.type == pygame.MOUSEBUTTONUP: #тышқан батырмасын босату
            work_surf.blit(screen, (0, 0)) #бетке суретті көшіру
            draw_on = False #сурет салуды тоқтату

        if event.type == pygame.MOUSEMOTION: #тышқанды жылжыту
            if draw_on: #егер сурет салу күйі қосылған болса
                if figure == 'pen': #қалам режимі
                    pygame.draw.line(screen, color, last_pos, event.pos, radius)
                    last_pos = event.pos 
                if figure == 'erase': #өшіргіш режимі
                    pygame.draw.circle(screen, (0, 0, 0), (event.pos[0], event.pos[1]), 6)
                if figure == 'rectangle': #тік төртбұрыш салу
                    t = Rect_pos(last_pos[0], last_pos[1], event.pos[0], event.pos[1])
                    screen.blit(work_surf, (0, 0))
                    pygame.draw.rect(screen, color, pygame.Rect(t))
                if figure == 'square': #шаршы салу
                    t = Square_pos(last_pos[0], last_pos[1], event.pos[0], event.pos[1])
                    screen.blit(work_surf, (0, 0))
                    pygame.draw.rect(screen, color, pygame.Rect(t))
                if figure == 'equilateral triangle': #тең қабырғалы үшбұрыш салу
                    t = Equilateral_Triangle_pos(last_pos[0], last_pos[1], event.pos[0], event.pos[1])
                    screen.blit(work_surf, (0, 0))
                    pygame.draw.polygon(screen, color, t)
                    points = []
                if figure == 'right triangle': #тік бұрышты үшбұрыш салу
                    t = Right_Triangle_pos(last_pos[0], last_pos[1], event.pos[0], event.pos[1])
                    screen.blit(work_surf, (0, 0))
                    pygame.draw.polygon(screen, color, t)
                    points = []
                if figure == 'rhombus': #ромб салу
                    t = Rhombus_pos(last_pos[0], last_pos[1], event.pos[0], event.pos[1])
                    screen.blit(work_surf, (0, 0))
                    pygame.draw.polygon(screen, color, t)
                    points = []
                if figure == 'circle': #шеңбер салу
                    screen.blit(work_surf, (0, 0))
                    pygame.draw.circle(screen, color, (last_pos[0], last_pos[1]), int(math.sqrt((event.pos[0]-last_pos[0])**2 + (event.pos[1]-last_pos[1])**2)))

    pygame.display.flip() #экранды жаңарту
pygame.quit() #pygame-ді жабу
```
